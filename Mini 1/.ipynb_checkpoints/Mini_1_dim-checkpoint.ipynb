```python
import sklearn.datasets
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_context('poster')
sns.set_palette('Paired', 10)
sns.set_color_codes()
```

```python
def benchmark_algorithm(dataset, dataset_sizes, cluster_function, function_args, function_kwds):

    # Initialize the result with NaNs so that any unfilled entries
    # will be considered NULL when we convert to a pandas dataframe at the end
    result = np.nan * np.ones((len(dataset_sizes)))
    labels = {}
    for index, size in enumerate(dataset_sizes):
            
            data = dataset[index]
            # Start the clustering with a timer
            start_time = time.time()
            k = cluster_function(data, *function_args, **function_kwds)
            time_taken = time.time() - start_time
            

            result[index] = time_taken
            try:
                labels[index] = k.labels_
            except:
                labels[index] = k
                
    # Return the result as a dataframe for easier handling with seaborn afterwards
    return result, labels
```

```python
import gensim
```

```python
text_file_name = 'wikipedia.txt'
```

```python
def create_word_vec(models, nork='n'):
    word_vecs = [model.wv for model in models]
    if nork=='n':
        data_sizes = [len(word_vect.vocab) for word_vect in word_vecs]
    else:
        data_sizes = [len(word_vect.syn0[0]) for word_vect in word_vecs]
    w2v = [dict(zip(word_vect.index2word,word_vect.syn0)) for 
           word_vect in word_vecs]
    return data_sizes, w2v
```

#### dim 50, lines 10000

```python
data_dim = data[1]
```

```python
dim_models_50 = [gensim.models.Word2Vec(data_dim, size=i, window=5,
                                      min_count=5, workers=8) for i in range(50,100,50)]
```

```python
k_sizes, w2v_large_k = create_word_vec(dim_models_50, 'k')
```

```python
w2v_large_k_vecs = [list(w.values()) for w in w2v_large_k]
```

```python
len(w2v_large_k_vecs[0][0])
```

```python
# 50 dimension
digits_proj_k = TSNE(random_state=1).fit_transform(w2v_large_k_vecs[0])
```

```python
k_means = sklearn.cluster.KMeans(10)
k_means_data, k_labels = benchmark_algorithm(w2v_large_k_vecs, k_sizes, k_means.fit, (), {})
```

```python
k_sizes[0]
```

```python
len(w2v_large_k_vecs[0])
```

```python
import random
indexes_50 = random.sample([i for i in range(len(w2v_large_k_vecs[0]))],400)
```

```python
plt.figure(figsize=(30,20))
plt.xlabel('Component #1')
plt.ylabel('Component #2')
all_points = [list(zip(list(w2v_large_k[0].keys()), digits_proj_k[:,0], digits_proj_k[:,1]))[i] for i in indexes_50]

for label, x, y in all_points:
        plt.annotate(label, xy=(x, y), xytext=(0, 0), size='small', textcoords='offset points')

plt.scatter([digits_proj_k[:,0][i] for i in indexes_50], [digits_proj_k[:,1][i] for i in indexes_50], 
            c=[k_labels[0][i] for i in indexes_50])
plt.show()
```

```python
dbscan = sklearn.cluster.DBSCAN(eps=0.2)
dbscan_data, db_labels = benchmark_algorithm(w2v_large_k_vecs, k_sizes, dbscan.fit, (), {})
```
